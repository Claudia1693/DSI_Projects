### 3.3 Plot `Verbal`, `Math`, and `Rate` appropriately on the same boxplot chart

```python
# Scale each variable of Verbal, Math and Rate along their min-max range respectively, so that each
# observation is now between 0 to 1, hence all three variables can be plotted on the same chart to
# see their distribution

verbal_min = float(df_sat['Verbal'].min())
verbal_range = df_sat['Verbal'].max() - verbal_min
math_min = float(df_sat['Math'].min())
math_range = df_sat['Math'].max() - math_min
rate_min = float(df_sat['Rate'].min())
rate_range = df_sat['Rate'].max() - rate_min

df_scaled = df_sat.copy()
for col in df_sat.columns:
    if col == 'Verbal':
        df_scaled[col].astype(float)
        for index, row in enumerate(df_scaled[col]):
            df_scaled.loc[index, col] = (df_sat.loc[index, col] - verbal_min)/verbal_range
    if col == 'Math':
        df_scaled[col].astype(float)
        for index, row in enumerate(df_scaled[col]):
            df_scaled.loc[index,col] = (df_sat.loc[index, col] - math_min)/math_range
    if col == 'Rate':
        df_scaled[col].astype(float)
        for index, row in enumerate(df_scaled[col]):
            df_scaled.loc[index,col] = (df_sat.loc[index, col] - rate_min)/rate_range   
```

```python
df_scaled.plot.box()
```

### 4.3 Assign and print the _covariance_ matrix for the dataset

```python
df_sat.cov()
```

Covariance is the sum of deviations from the mean of first variable multiplied by deviations from mean of second variable divided by sample size. However, covariance is difficult to read and interprete because it is relative to the variance of the variables.   

Covariance can be converted to correlation by dividing covariance by standard deviation of first variable and standard deviation of second variable.  

Correlation is preferred because it is dimensonless and between the values of -1 to 1, which is easy to interprete.  
A value of -1 indicates a perfect negative correlation,  
a value of 1 is a perfect positive correlation while  
a value of 0 means no correlation at all.

## 5. Performing EDA on "drug use by age" data.

---

### 5.1 Load the data, clean the data and check for missing values.

```python
drugs = pd.read_csv('drug-use-by-age.csv')
```

```python
drugs.head()
```

```python
drugs.info()
```

From the .info() method, the following fields should be numeric but are not, hence require cleaning:  
- cocaine-frequency, 
- crack-frequency, 
- heroin-frequency, 
- inhalant-frequency, 
- oxycontin-frequency, 
- meth-frequency

```python
# To replace the '-' character with the numeric zero and convert the columns to data type float

to_clean = ['cocaine-frequency', 'crack-frequency', 'heroin-frequency', 'inhalant-frequency',
            'oxycontin-frequency', 'meth-frequency']

for label in to_clean:
    drugs.loc[drugs[label]=='-', label] = 0.
    drugs[label] = pd.to_numeric(drugs[label], downcast='float')
```

```python
drugs.describe()
```

### 5.2 Do a high-level, initial overview of the data


```python
# Split the original drugs table into 2 sub-tables:
# 1. drugs_users - lists the percentage of drug users per age group
# 2. drugs_usage - lists the median number of times a drug user used each drug in the past 12 months

drugs_users_percent = drugs.copy()
drugs_usage = drugs.copy()

for header in drugs_users_percent.columns:
    if 'frequency' in header:
        drugs_users_percent = drugs_users_percent.drop(columns=header)
    elif 'use' in header:
        drugs_usage = drugs_usage.drop(columns=header)
```

```python
x = drugs_usage['age']

fig = plt.figure(figsize=(10,10))

for col in drugs_usage.columns[2::]:
    y = drugs_usage[col]
    plt.plot(x,y)

plt.title('Use frequency of each drug across age')
plt.xlabel('age')
plt.ylabel('usage')
plt.legend(loc='best')
plt.show()
```

```python
x = drugs_users_percent['age']

fig = plt.figure(figsize=(10,10))

for col in drugs_users_percent.columns[2::]:
    y = drugs_users_percent[col]
    plt.plot(x,y)

plt.title('Percentage users of each drug across age')
plt.xlabel('age')
plt.ylabel('usage')
plt.legend(loc='best')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(drugs_users_percent.corr())
```

The heatmap indicates a strong correlation in the number of drug users by age group among the following drugs:
1. pain-reliever
2. oxycontin
3. tranquilizer
4. stimulant
5. meth
6. marijuana
7. cocaine

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(drugs_usage.corr())
```

From the heatmap, there is correlation in drug usage between the following drug categories:
1. marijuana and alcohol
2. crack and cocaine
3. inhalant and hallucinogen

### 5.3 Create a testable hypothesis about this data

**T-test**  
Null hypothesis: that the drug users of marijuana and of oxycontin are from the same population  
Alternative hypothesis: that the drug users of the two drugs are not from the same population

```python
sample1 = drugs_users['marijuana-use']
sample2 = drugs_users['oxycontin-use']

stats.ttest_ind(sample1, sample2)

```

As the p-value is below 0.05, the null hypothesis can be rejected and the drug users of marijuana are statistically significant from the users of oxycontin 

## 6. Introduction to dealing with outliers

---

Assume definition of outliers lies 1.5 times of interquartile range from the lower and upper quartile

```python
iqr = df_sat['Rate'].quantile(.75) - df_sat['Rate'].quantile(.25)

rate_lower = df_sat['Rate'].quantile(.25) - iqr

# Since there can't be negative rate, if the lower range of rate is negative, set it to zero
if rate_lower < 0:
    rate_lower = 0
    
rate_upper = df_sat['Rate'].quantile(.75) + iqr

rate_outliers = df_sat[(df_sat['Rate']<rate_lower) | (df_sat['Rate']>rate_upper)]
rate_outliers
```

```python
df_sat['Rate'].hist(bins=20)
```

There are no outliers, however from the histogram plot, it appears that there might be 2 clusters of states with 2 distinct distributions of the participation rate
